### 2

```python
dfLawSum = dfSortedDP.groupby(['lawArticle']).size().reset_index(name="deathPenaltyCount")
dfP = dfSortedDP.groupby(['province_EN', 'lawArticle']).size().reset_index(name="deathPenaltyCount")
```

```python
dfLawSum
```

```python
lawDict = dfLawSum.set_index('lawArticle')['deathPenaltyCount'].to_dict()
```

```python
newArticleSList = []
newDPlist = []
dfLawTolist = dfP['lawArticle'].tolist()
dfCountToList = dfP['deathPenaltyCount'].tolist()

for idx, art in enumerate(dfLawTolist):
    newArticleSList.append('law_{}'.format(art))
    newDPlist.append(dfCountToList[idx]/lawDict[art])
dfP['lawArticleS'] = newArticleSList
dfP['deathPenaltyPercentage'] = newDPlist
```
